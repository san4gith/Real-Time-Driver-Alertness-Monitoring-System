```python
import matplotlib.pyplot as plt
import numpy as np

# Create the figure and axis
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Generate the data
num_shelves = 5
num_books_per_shelf = 8
shelf_height = 0.5
book_width = 0.2
book_depth = 0.1

# Create the bookshelf coordinates
x_shelf = np.repeat(np.arange(num_shelves), num_books_per_shelf)
y_shelf = np.tile(np.arange(num_books_per_shelf), num_shelves)
z_shelf = np.zeros(num_shelves * num_books_per_shelf)

# Plot the bookshelves
ax.bar3d(x_shelf, y_shelf, z_shelf, 1, 0.1, shelf_height, color='#9575CD')

# Set the view angle and adjust spacing
ax.view_init(elev=20, azim=45)
ax.dist = 10

# Remove axis grids
ax.grid(False)
ax.axis('off')

# Display the plot
plt.show()
```
